```python
#The program prompts user to choose a number for playing the game for n number of times. 
#after that the program should prompt user to choose a number between 1 to 26.
#The picked number should be between 1 to 26 and it should match the random number generated by random module (package) . 
#The console will print about whether the picked number is greater or smaller than the generated random number. 
#The process will continue till it match the random number and the console will print about the number of attempts it took to do this.
```

```python
import random
name=str(input("what's your name?  "))
print("good luck", name)
m=True
while m:
    n=random.randint(1,26)
    c=0
    g=int(input("enter yout guess game between 1 to 26: "))
    while g!=n:
        c=c+1
        if g>n+5:
            print("guess is high")
        elif g<n-5:
            print("guess is low")
        elif g>n:
            print("little high")
        elif g<n:
            print("little low")
        g=int(input("try again: "))
    print("you guess the no in: " ,c,"attemps")
    rem=str(input("play again: "))
    if rem=="no":
        m=False
```
